```python
#input words
input_words = input()
input_list = input_words.split() #converting input words into list of words
print(input_list)
```

```python
#funtion for tranforming data from text files into list of words
def txt_to_list(file_name):
  entry_list = []
  import string
  import codecs
  with codecs.open(file_name, 'r',encoding='utf-8',errors='ignore') as f: #reading text files
    lines = f.readlines()
    for i in lines:
      for k in i.split():
        entry_list.append(k.lower()) #converting words to lower case and adding them to list
    
  listname = []

  for m in entry_list:
    n = m.translate(str.maketrans('', '', string.punctuation)) #removing punctuations
    listname.append(n)
  return listname
```

```python
list_dict = {} #creating a dictionary with all text file words as value lists

for i in range(1,21):
  if i < 10:
    list_dict['l'+str(i)] = txt_to_list('/' + '0'+str(i)+'.txt.txt') #using txt_to_list function
  else:
    list_dict['l'+str(i)] = txt_to_list('/' + str(i)+'.txt.txt')

print(list_dict)
```

```python
#creating list of stop words
stop_words = []

import string

with open('/stopwords.txt', 'r') as f:
  lines = f.readlines()
  for i in lines:
    for k in i.split():
      stop_words.append(k.lower())#converting stop words to lower case ans adding them to list
  
final_stop_words = []

for m in stop_words:
  n = m.translate(str.maketrans('', '', string.punctuation))#removing punctuations from stop words for uniformity
  final_stop_words.append(n)

print(final_stop_words)
```

```python
#removing stop words from text file words
final_list_dict = {} 
for i in range(1,21):
  final_list_dict['fl'+str(i)] = [i for i in list_dict['l'+str(i)] if i not in final_stop_words]#final lists of words after removing stop words

print(final_list_dict)
```

```python
#creating a list with all words from text files after removing stop words
all_list = []

for i in range(1,21):
  all_list += final_list_dict['fl'+str(i)]#list that includes words from all text file after removing stop words

print(all_list)
```

```python
#mapping function
def map(id,lst):
  map_op = []
  for word in lst:
    tmp = (word,1)
    map_op.append(tmp)
  return map_op
```

```python
#output dictionary
final_output = {}
```

```python
#reducing function
def reduce(n,lst):

  for i in range(n):
    final_output['doc'+str(i)]={} #creates nested dictionaries based on number of documents

  for i in range(n):
    for k,v in lst[i]:
      if k in input_list: #checks for words in input list of words
        if k not in final_output['doc'+str(i)]:
          final_output['doc'+str(i)][k]=0
        final_output['doc'+str(i)][k] += v #gives count based on key
  return final_output
```

```python
#main
if __name__ == "__main__":
  #mapping process
  map_list = [all_list]
  for i in range(1,21):
    map_list += [final_list_dict['fl'+str(i)]]#creates lists with nested lists
  final_map_output = []
  for i in range(21):
    map_output = map(i,map_list[i]) #using mapping function
    final_map_output += [map_output]
  print(final_map_output)

  reduce(21,final_map_output) #using reducing funtion with input list of word

  print(final_output)#gives count of given input words in all files and in each file
```

```python
#removes none dictionaries from final output and stores as dictionary final results
final_results = {key : val for key, val in final_output.items() if val}
print(final_results)
```

```python
#final output printing
for i in final_results:
  if i == 'doc0':
    print(f"{input_words} repeated {sum(final_results[i].values())} times")#gives count of input words in all files combined

for i in final_results:
  if i != 'doc0':
    for k,v in final_results[i].items():
      print(f"{k} repeated {v} times in {i}")#gives count of input words in invidual docs
```
